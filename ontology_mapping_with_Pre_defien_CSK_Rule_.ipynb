```python
def update_ontology_with_class_and_instance(ontology, class_name, parent_class_iri):
    print(f"Creating class: {class_name} under parent IRI: {parent_class_iri} in ontology: {ontology}")

    # Check if the class already exists
    existing_class = ontology.search(iri="*" + class_name)
    if not existing_class:
        parent_class = ontology[parent_class_iri]
        if parent_class is None:
            print(f"Parent class with IRI '{parent_class_iri}' not found in ontology.")
            return None, None

        with ontology:
            NewClass = types.new_class(class_name, (parent_class,))
            instance = NewClass(class_name.replace(" ", "_"))
            return NewClass, instance
    else:
        return existing_class[0], existing_class[0](class_name.replace(" ", "_"))
```

```python
#RuleTemplate Class: Represents a rule template, which contains a logical expression with placeholders.

class RuleTemplate:
    def __init__(self, expression: str):
        """Initialize the RuleTemplate class with an expression.

        Args:
        expression (str): A logical expression containing placeholders.
        """
        self.expression = expression

    def __str__(self):
        return self.expression
```

```python
#CSK Class: Represents a Manufacturing Common Sense Knowledge statement, which provides specific knowledge about painitng
class CSK:
    def __init__(self, statement: str):
        """Initialize the CSK class with a statement.

        Args:
        statement (str): A statement that provides specific knowledge about painting.
        """
        self.statement = statement

    def __str__(self):
        return self.statement


```

```python
#ConcreteRule Class: Represents a concrete rule derived from a rule template by replacing its placeholders with specific classes/instances.
class ConcreteRule:
    def __init__(self, expression: str):
        """Initialize the ConcreteRule class with an expression.

        Args:
        expression (str): A logical expression that's the result of substituting placeholders in a RuleTemplate with information from an CSK.
        """
        self.expression = expression

    def __str__(self):
        return self.expression

```

```python
from owlready2 import *

def update_ontology_with_class_and_instance(ontology, class_name, parent_class_iri, new_ontology):
    csk_prefix = "http://www.CSKG.enit.fr"  # Custom prefix for new classes and instances
    new_class_iri = f"{csk_prefix}#{class_name}"

    
    parent_class = new_ontology.search_one(iri=parent_class_iri) # Retrieve or create the parent class in the new ontology
    if not parent_class:
        # Handle the case where the parent class doesn't exist
        raise ValueError(f"Parent class with IRI '{parent_class_iri}' not found in the ontology.")

    
    if new_class_iri in new_ontology.classes():# Create or get the class in the new ontology
        ont_class = new_ontology[new_class_iri]
    else:
        
        with new_ontology:
            ont_class = types.new_class(class_name, (parent_class,))# Create a new class under the parent class in the new ontology
            ont_class.iri = new_class_iri

    # Create an instance in the new ontology
    instance_name = f"{class_name.lower()}_instance"
    instance_iri = f"{csk_prefix}#{instance_name}"
    with new_ontology:
        ont_instance = ont_class(name=instance_name, namespace=new_ontology)
        ont_instance.iri = instance_iri

    return ont_class, ont_instance



# Step 1: Commonsense Driven Semantic Rule Templates

RT1 = RuleTemplate("∀x (product(x) → ∃y (process(y) ∧ isoutputOf(x, y)))")
RT2 = RuleTemplate("∀x (process(x) → ∃y (process(y) ∧ precedes(x, y)))")
RT3 = RuleTemplate("∀x (process(x) → ∃y (machine(y) ∧ participatesAtSomeTime(y, x)))")

def determine_rule_template(csk: CSK) -> RuleTemplate:
    if "result" in csk.statement:
        return RT1
    elif "After" in csk.statement:
        return RT2
    elif "involves" in csk.statement:
        return RT3
    else:
        raise ValueError("Unknown CSK format!")

# ... [Class Definitions for RuleTemplate, CSK, and ConcreteRule] ...

def SpecializeRule(RT: RuleTemplate, CSK: CSK) -> ConcreteRule:
    words = CSK.statement.split()

    # Handle RT1
    if "result" in CSK.statement:
        process_name = words[3]
        product_name = ' '.join(words[-3:]).replace('a ', '').replace('an ', '').rstrip('.')
        process_class, process_instance = update_ontology_with_class_and_instance(csk_onto, process_name, "http://purl.obolibrary.org/obo/BFO_0000015", csk_onto)
        product_class, product_instance = update_ontology_with_class_and_instance(csk_onto, product_name, "http://purl.obolibrary.org/obo/BFO_0000040", csk_onto)
        CR_expression = f"∀x ({process_instance.name}(x) → ∃y ({product_instance.name}(y) ∧ isOutputof,IRI:https://spec.industrialontologies.org/ontology/core/Core/isOutputOf (y, x)))"

    # Handle RT2
    elif "After" in CSK.statement:
        preceding_process = words[1]
        succeeding_process = words[4]
        preceding_process_class, preceding_process_instance = update_ontology_with_class_and_instance(csk_onto, preceding_process, "http://purl.obolibrary.org/obo/BFO_0000015", csk_onto)
        succeeding_process_class, succeeding_process_instance = update_ontology_with_class_and_instance(csk_onto, succeeding_process, "http://purl.obolibrary.org/obo/BFO_0000015", csk_onto)
        CR_expression = f"∀x ({preceding_process_instance.name}(x) → ∃y ({succeeding_process_instance.name}(y) ∧ BFO_0000063, IRI: http://purl.obolibrary.org/obo/BFO_0000063 (x, y)))"


    # Handle RT3
    elif "involves" in CSK.statement:
        process_name = words[1]
        machine_name = ' '.join(words[-2:]).replace('a ', '').replace('an ', '').rstrip('.')
        process_class, process_instance = update_ontology_with_class_and_instance(csk_onto, process_name, "http://purl.obolibrary.org/obo/BFO_0000015", csk_onto)
        machine_class, machine_instance = update_ontology_with_class_and_instance(csk_onto, machine_name, "https://spec.industrialontologies.org/ontology/core/Core/MaterialArtifact", csk_onto)
        CR_expression = f"∀x ({process_instance.name}(x) → ∃y ({machine_instance.name}(y) ∧ BFO_0000056, IRI: http://purl.obolibrary.org/obo/BFO_0000056,(y, x)))"

    return ConcreteRule(CR_expression)

# ... [ determine_rule_template, generate_concrete_rule, and main...]

def generate_concrete_rule(csk_input: str) -> ConcreteRule:
    csk = CSK(csk_input)
    rt = determine_rule_template(csk)
    return SpecializeRule(rt, csk)

def main():
    print("Enter CSK statements (one per line). Type 'END' to stop:")
    csk_inputs = []
    while True:
        line = input()
        if line == 'END':
            break
        csk_inputs.append(line)

    for csk_input in csk_inputs:
        try:
            cr = generate_concrete_rule(csk_input)
            print(f"Generated Concrete Rule for '{csk_input}': {cr}")
        except ValueError as e:
            print(f"Failed to generate rule for '{csk_input}': {e}")

main()

# Save the updated ontologies
csk_onto.save(file="test.rdf", format="rdfxml")

```
